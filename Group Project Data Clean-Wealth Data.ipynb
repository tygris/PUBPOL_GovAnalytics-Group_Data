## Pull subset of data to be used

```python
controls1 = wealth.iloc[0:217, [0,1,3,4]]
controls1.head()
```

```python
controls1.tail()
```

## Clean column names

```python
#use dictionary for easy updating later
OldToNew={controls1.columns[0]:'Country',
          controls1.columns[1]:'Population',
          controls1.columns[2]:'GDP_USdol',
          controls1.columns[3]:'GDPperCap_USdol'
         }
#set the new column names
controls1.rename(columns=OldToNew,inplace=True)
controls1.head()
```

## Clean the data

```python
# get list of columns that have numeric values
numericCols = controls1.iloc[:,1:].columns
# remove the commas from all numeric values
for i in numericCols:
    controls1[i] = controls1[i].str.replace(',', '')

controls1.head()
```

```python
badSymbols=[] # list for bad symbols

# get list of cells that need to be converted to NaNs
for columnName in numericCols:
    for cell in controls1[columnName]:
        try:
            float(cell) 
        except:             
            if cell not in badSymbols:               
                badSymbols.append(cell)

# fix the cell values:
controls1.replace(to_replace=badSymbols,value=np.nan,inplace=True)
controls1
```

## Set the data types

```python
# set desired data types for the columns
for col in numericCols:
    controls1[col] = controls1[col].apply(pd.to_numeric)
    
controls1.dtypes
```
